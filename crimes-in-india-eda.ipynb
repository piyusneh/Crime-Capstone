# Data Analysis on Rape Victims of India, of 2016 & 2018

```r
library(tidyverse)
library(reshape2)
require(gridExtra)
library(ggthemes)
list.files(path = "../input")
```

```r
vrape16 = read_csv('../input/crime-in-india/Victims of Rape/2016 Victims of Rape.csv')
vrape18 = read_csv('../input/crime-in-india/Victims of Rape/2018 Victims of Rape.csv')
vrape16
vrape18
```

```r
names(vrape16) = gsub(" ", "", names(vrape16))   #Cleaning Columns Names
names(vrape16) = gsub("/", "", names(vrape16))
names(vrape16) = gsub("-", "", names(vrape16))
names(vrape16) = gsub('\\(', "", names(vrape16))
names(vrape16) = gsub('\\)', "", names(vrape16))
names(vrape16) = gsub('\\&', "", names(vrape16))


names(vrape18) = gsub(" ", "", names(vrape18))
names(vrape18) = gsub("/", "", names(vrape18))
names(vrape18) = gsub("-", "", names(vrape18))
names(vrape18) = gsub('\\(', "", names(vrape18))
names(vrape18) = gsub('\\)', "", names(vrape18))
names(vrape18) = gsub('\\&', "", names(vrape18))

options(repr.plot.width = 17, repr.plot.height = 9)
```

# Top 10 States of Rape Victim

```r
P3 = vrape16%>%
arrange(desc(TotalVictims))%>%
head(10)%>%
ggplot(aes(x=reorder(StateUT,TotalVictims), y=TotalVictims,fill = '#8c1aff'))+geom_col(show.legend = FALSE)+coord_flip()+
geom_label(aes(x=reorder(StateUT,TotalVictims), y=TotalVictims,fill = '#8c1aff', label = TotalVictims),fill = 'white',show.legend = FALSE)+
labs(title = 'Top 10 States according to #Rape Victims: India',
     subtitle = 'Top 10 States (2016)',
    y= 'No of Victims',
    x='Top 10 States',
    caption = 'Data Source: Kaggle')+
theme_fivethirtyeight()
P4 = vrape18%>%
arrange(desc(TotalVictims))%>%
head(10)%>%
ggplot(aes(x=reorder(StateUT,TotalVictims), y=TotalVictims,fill = '#8c1aff'))+geom_col(show.legend = FALSE)+coord_flip()+
geom_label(aes(x=reorder(StateUT,TotalVictims), y=TotalVictims,fill = '#8c1aff', label = TotalVictims),fill = 'white',show.legend = FALSE)+
labs(title = 'Top 10 States according to #Rape Victims: India',
     subtitle = 'Top 10 States (2018)',
    y= 'No of Victims',
    x='',
    caption = 'Data Source: Kaggle')+
theme_fivethirtyeight()+
theme(axis.text.y=element_text(size = 10))

grid.arrange(P3,P4, ncol=2)
```

# Composition of Adult and Non Adult Victims For the Top 10 affected States. 

```r
df1 = vrape16%>%
transmute(Total_non_audult = ChildVictimsofRapeBelow18YrsTotalGirlChildVictims,
         Total_audult = WomenVictimsofRapeAbove18YrsTotalWomenAdultVictims,
         state = StateUT,
         Total = TotalVictims)%>%   
arrange(desc(Total))%>%
head(10)
#Transmute for Having only the new columns

dfm = melt(df1[,1:3],id.vars = 3)

P5 = ggplot(dfm,aes(x = state,y = value,fill = variable)) + 
    geom_bar(stat = "identity",position = "stack")+
labs(title = 'Composition of Adult and Minor Victims',
     subtitle = 'Top 10 States; Total Victims (2016):India',
    y= 'No of Victims',
    x='States',
    caption = '')+
geom_text(aes(x = state,y = value,label = value),
          position = position_stack(vjust = .5),
            colour = "black", size = 5)+
theme_wsj()+
theme(axis.text.x = element_text(margin=margin(10, b = 10),angle = 90),
     axis.text.y = element_text(color = 'red'),
     legend.title = element_blank())


df2 = vrape18%>%
transmute(Total_non_audult = ChildVictimsofRapeBelow18YrsTotalGirlChildVictims,
         Total_audult = WomenVictimsofRapeAbove18YrsTotalWomenAdultVictims,
         state = StateUT,
         Total = TotalVictims)%>%
arrange(desc(Total))%>%
head(10)

dfm = melt(df2[,1:3],id.vars = 3)

P6 = ggplot(dfm,aes(x = state,y = value,fill = variable)) + 
    geom_bar(stat = "identity",position = "stack")+
labs(title = 'Composition of Adult and Minor Victims',
     subtitle = 'Top 10 States; Total Victims (2018):India',
    y= 'No of Victims',
    x='States',
    caption = '')+
geom_text(aes(x = state,y = value,label = value),
          position = position_stack(vjust = .5),
            colour = "black", size = 5)+
theme_wsj()+
theme(axis.text.x = element_text(margin=margin(10, b = 10),angle = 90),
     axis.text.y = element_text(color = 'red'),
     legend.title = element_blank())
#Arranging Two Plots side by side
grid.arrange(P5,P6, ncol=2)
```

From the above diagram it is clearthat that in both the year 2016 and 2018 Madhya Pardesh suffered the highest number of victims. Uttar Prdesh and Rajsthan are the second and third affected state in both the year and they just exchanged place with one another through out the year. Number of Total Non Adult Victims decreased and the counter part increased For Madhya Pradesh from 2016 to 2018. States like Delhi and Maharashtra and Assam has registered a significantly lower number of Non Adult Victims in the list of Top 10 states. 

# Difference between #Victim and #Case Reported

```r
set.seed(25)
P1= vrape16%>%
mutate(dif1 = TotalVictims-CasesReported )%>%
arrange(desc(dif1))%>%
head(10)%>%
ggplot(aes(x=reorder(StateUT,dif1), y=dif1,fill = '#ff4d4d'))+geom_col(show.legend = FALSE)+coord_flip()+
geom_label(aes(x=reorder(StateUT,dif1), y=dif1,fill = '#ff4d4d', label = dif1),fill = 'white',show.legend = FALSE)+
labs(title = 'Difference in #Rape Victim and Reported Case: India',
     subtitle = 'Top 10 States (2016)',
    y= 'No of Differences',
    x='Top 10 States',
    caption = 'Data Source: Kaggle')+
theme_pander()


P2 = vrape18%>%
mutate(dif3 = TotalVictims-CasesReported )%>%
arrange(desc(dif3))%>%
head(10)%>%
ggplot(aes(x=reorder(StateUT,dif3), y=dif3,fill = '#ff4d4d'))+geom_col(show.legend = FALSE)+coord_flip()+
geom_label(aes(x=reorder(StateUT,dif3), y=dif3,fill = '#ff4d4d', label = dif3),fill = 'white',show.legend = FALSE)+
labs(title = 'Difference in #Rape Victim and Reported Case :India',
     subtitle = 'Top 10 States (2018)',
    y= 'No of Differences',
    x='Top 10 States',
    caption = 'Data Source: Kaggle')+
theme_pander()


grid.arrange(P1,P2, ncol=2)
```

This particular indicator could be a point of concern for the law and Administration for the particular state. This shows the difference between the no. of Victims and the total case reported with Police. The greater the number is the worse the situation of Law and Administration of the Particular State. Uttar Pradesh is in the top for both the years. Telangana Performed well and from 3rd Position in 2016 the states is not in top 10 in 2018. The allover status of India has also became better in 2018. 

# Comparison of Age Distribution of Rape Victim; Top Four States 2016 

```r
library(grid)
rvrape = vrape16%>%
arrange(desc(TotalVictims))%>%
rename( State_UT = StateUT,
           Case_reported = CasesReported,
          Below_6 = ChildVictimsofRapeBelow18YrsBelow6Years,
          Below12_above6 = ChildVictimsofRapeBelow18Yrs6YearsAboveBelow12Years,
         Below16_above12 = ChildVictimsofRapeBelow18Yrs12YearsAboveBelow16Years,
         Below18_above16 = ChildVictimsofRapeBelow18Yrs16YearsAboveBelow18Years,
      Total_NA = ChildVictimsofRapeBelow18YrsTotalGirlChildVictims,
      Below30_above18 = WomenVictimsofRapeAbove18Yrs18YearsAboveBelow30Years,
      Below45_above30 = WomenVictimsofRapeAbove18Yrs30YearsAboveBelow45Years,
      Below60_above45 = WomenVictimsofRapeAbove18Yrs45YearsAboveBelow60Years,
      Above60 = WomenVictimsofRapeAbove18Yrs60YearsAbove,
      Total_A = WomenVictimsofRapeAbove18YrsTotalWomenAdultVictims)
P7 = rvrape%>%
filter(State_UT == 'Madhya Pradesh')%>%
melt(id.vars = 3)%>%
filter(!value == 'State')%>%
filter(!variable == 'Case_reported')%>%
filter(!variable == 'S.No',
      !variable == 'Total_NA',
      !variable == 'Total_A',
      !variable == 'TotalVictims')%>%
#arrange(desc(as.numeric(value)))
ggplot(aes(x=variable,y=as.numeric(value)))+
geom_col(fill = '#e60000')+
labs(title = '1. Madhya Pradesh',
    x= 'Age Group',
    y='#Victims',
    caption = '')+
theme_wsj()+
theme(axis.text.x = element_text(size = 7.3),
     axis.text.y = element_text(color = 'red'))

P8 = rvrape%>%
filter(State_UT == 'Uttar Pradesh')%>%
melt(id.vars = 3)%>%
filter(!value == 'State')%>%
filter(!variable == 'Case_reported')%>%
filter(!variable == 'S.No',
      !variable == 'Total_NA',
      !variable == 'Total_A',
      !variable == 'TotalVictims')%>%
#arrange(desc(as.numeric(value)))
ggplot(aes(x=variable,y=as.numeric(value)))+geom_col(fill = '#e60000')+
labs(title = '2. Uttar Pradesh',
    x= 'Age Group',
    y='#Victims',
    caption = '')+
theme_wsj()+
theme(axis.text.x = element_text(size = 7.3),
     axis.text.y = element_text(color = 'red'))


P9 = rvrape%>%
filter(State_UT == 'Rajasthan')%>%
melt(id.vars = 3)%>%
filter(!value == 'State')%>%
filter(!variable == 'Case_reported')%>%
filter(!variable == 'S.No',
      !variable == 'Total_NA',
      !variable == 'Total_A',
      !variable == 'TotalVictims')%>%
#arrange(desc(as.numeric(value)))
ggplot(aes(x=variable,y=as.numeric(value)))+geom_col(fill = '#e60000')+
labs(title = '3. Rajasthan',
    x= 'Age Group',
    y='#Victims',
    caption = '')+
theme_wsj()+
theme(axis.text.x = element_text(size = 7.3),
     axis.text.y = element_text(color = 'red'))

P10 = rvrape%>%
filter(State_UT == 'Odisha')%>%
melt(id.vars = 3)%>%
filter(!value == 'State')%>%
filter(!variable == 'Case_reported')%>%
filter(!variable == 'S.No',
      !variable == 'Total_NA',
      !variable == 'Total_A',
      !variable == 'TotalVictims')%>%
#arrange(desc(as.numeric(value)))
ggplot(aes(x=variable,y=as.numeric(value)))+geom_col(fill = '#e60000')+
labs(title = '4. Odisha',
    x= 'Age Group',
    y='#Victims',
    caption = 'Data Source: Kaggle')+
theme_wsj()+
theme(axis.text.x = element_text(size = 7.3),
     axis.text.y = element_text(color = 'red'))

grid.arrange(P7,P8,P9,P10, ncol=2,nrow=2,
             top = textGrob("No of Rape Victims according to Age Group, Top four States of India (2016)",
                            gp=gpar(fontsize=25,fontface = 'bold',fontsize=25)))
```

Madhya Pradesh which is in the top of the list shows a disturbing number for Non Adult Victims of Rape. Specially the age group of Above 12 and Below 16 suffered a lot in comparison with other top 4 states in the year 2016. For the Second and Third state the concerning age group is Below 30 and above 18 years of age. And this also point of Concern for Madhya Pradesh too. In Odisha the age group of 16 to 18 years of age is affected the most.
